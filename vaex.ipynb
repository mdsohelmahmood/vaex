```python
# convert csv to hsf5 format
dv = vaex.from_csv('BIGFILE.csv', convert=True, chunk_size=5_000_000)
```

```python
dv.plot(dv.col1, dv.col2)
```

```python
# virtual column: Virtual columns behave just like normal ones, but they take up no memory 
# what so ever. This is because Vaex only remembers the expression the defines them, and 
# does not calculate the values up front. These columns are lazily evaluated only when necessary, 
# keeping memory usage low.


dv['col1/2']=dv.col1/2
dv['col1/2']
```

```python
dv.col1.value_counts()
```

```python
dv_50=dv[dv.col1>50]
```

```python
dv_50
```

```python
dv_50.col1.value_counts()
```

```python
# "selection" feature of vaex. Use of statistics on the selected data

# dv.col1.mean()
dv.col1.mean(selection=[None, dv.col1>50, dv.col1<50])
```

```python
dv.col1.std(selection=[None, dv.col1>50, dv.col1<50])
```

```python
# groupby feature of vaex with aggregations

dv.groupby(by='Group', agg={'count':vaex.agg.count()})

```
